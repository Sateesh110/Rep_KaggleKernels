### Null Records Analysis

```python
# columns
columns = df.columns
```

```python
# df stats function

def df_stats(df,columns):
    '''
    Stats of each column:
    1.Count
    2.Unique Values count
    3.Missing Values count 
    4.% of Missing Values wrt total instances of the dataset
    5.% of Highest column value count
    6.Data Type
    
    '''
    
    stats = []
    for col in columns:
        stats.append((col, df[col].count(), df[col].nunique(), df[col].isnull().sum(), df[col].isnull().sum() * 100 / df.shape[0], df[col].value_counts(normalize=True, dropna=False).values[0] * 100, df[col].dtype))
    
    stats_cols = ['feature', 'record_count', 'unique_values','missing_records', 'percent_missing_records','percent_highest_columnvalue', 'data_type']
    df_stats = pd.DataFrame(stats, columns=stats_cols)
    df_stats['cat_or_num'] = df_stats['data_type'].apply(lambda x: 'categorical' if x == 'object' else 'numerical')
    return(df_stats) 
```

```python
# run the above function
df_stats = df_stats(df,columns)
```

```python
# Check for Missing Values
df_stats.sort_values('percent_missing_records', ascending=False)
```

```python
# Visualization of Missing Values as well as % of biggest column value
sns.relplot(y="feature", x="percent_highest_columnvalue", size="percent_highest_columnvalue", sizes=(0,60),style="cat_or_num",hue="percent_missing_records",  data=df_stats, height=8.27, aspect=11.7/8.27)
```

### Create a Copy

```python
# create a copy of the data 
df2= df.copy()
```

### Check for duplicate instances

```python
df2 = df2.drop_duplicates(['animal_id'], keep ='last')
```

```python
df2['animal_id'].count()
```

### Drop Columns (Nulls > 75%)

```python
# drop columns which have missing records > 75% 
drop_cols = []
for col in columns:
    rate = (df2[col].isnull().sum() * 100 / df2.shape[0])
    if rate > 75:
        df2 = df2.drop(col, axis=1)
        drop_cols.append(col)
        
print('dropped columns:', drop_cols)
print('rows, columns',df2.shape)
```

### Convert String to Date

```python
df2['date_of_birth'] = pd.to_datetime(df2['date_of_birth'])

df2['dob_year'] = df2['date_of_birth'].dt.year
df2['dob_month'] = df2['date_of_birth'].dt.month
df2['dob_monthyear'] = pd.to_datetime(df2['monthyear']).dt.to_period('M')

df2['datetime'] = pd.to_datetime(df2['datetime'])
df2['outcome_month'] = df2['datetime'].dt.month
df2['outcome_year'] = df2['datetime'].dt.year
df2['outcome_weekday'] = df2['datetime'].dt.weekday_name
df2['outcome_hour'] = df2['datetime'].dt.hour
```

### New Columns

#### Sex

```python
df2['sex_upon_outcome'][0:5]
```

```python
df2['sex'] = np.where(df2['sex_upon_outcome'].str.contains('Male'), 'Male', 'Female')
df2['sex'][0:5]
```

#### spay_neuter

```python
df2['spay_neuter'] = np.where(df2['sex_upon_outcome'].str.contains('Intact'), 'No', 'Yes')
```

#### Split: Color1,Color2

```python
df2['color1'], df2['color2'] = df2['color'].str.split('/', 1).str
```

#### Replace /: Color3

```python
df2['color3']= df2['color'].str.replace('/','-')
```

#### Split : Period, Range

```python
df2['period'], df2['range'] = (df2['age_upon_outcome'].str.split(' ').str[0].fillna(0).astype(int)), (df2['age_upon_outcome'].str.split(' ').str[1].fillna(0))
```

```python
df2['range'].unique()
```

#### Range2, Range3

##### Option1

```python
df2['range2'] = np.where(df2['range'].str.contains('day'),1,
                            np.where(df2['range'].str.contains('week'),7,
                               np.where(df2['range'].str.contains('month'),30,
                                  np.where(df2['range'].str.contains('year'),365,0)))).astype(int)
                                   
```

##### Option2

```python
# Create a function

def range_conversion(range_value):
    if range_value in ('day','days'):
        return 1
    elif range_value in ('week','weeks'):
        return 7
    elif range_value in ('month','months'):
        return 30
    elif range_value in ('year','years'):
        return 365
    else:
        return 0        
```

```python
# apply function
# Use always option 2
df2['range3'] = df2['range'].apply(range_conversion)
```

```python
query = (df2['range']==0)
df2[query][0:2]
```

#### age_days, age_years

```python
df2['age_days'] = df2['range3'] * df2['period']
df2['age_years'] = df2['age_days'] / 365
```

#### cat_kitten

```python
df2['cat_kitten'] = np.where(df2['age_days'] >= 180, 'Cat', 'Kitten')
```

```python
df2['adult_child'] = df2['age_days'].apply(lambda x:'Adult' if x >=180 else 'Child')
```

```python
df2[['cat_kitten','adult_child']][0:3]
```

#### Age_Group

```python
df2['age_group'] = pd.cut(df2['age_years'], 10)
df2['age_group'].value_counts()
```

#### Breed: Replace Values 

```python
# Replace Mix with " "
# Replace 'medium hair', 'mediumhair'

df2['breed'] = df2['breed'].str.lower().str.replace('mix', '').str.replace('medium hair', 'mediumhair').str.strip()
```
